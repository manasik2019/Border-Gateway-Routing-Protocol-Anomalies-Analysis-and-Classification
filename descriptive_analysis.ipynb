```python
#BGP Features Trend before, during and after th edifferent anomalies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

WannaCrypt=pd.read_csv('WannaCrypt.csv')
Moscow_blackout = pd.read_csv('Moscow_blackout1.csv')
Nimda=pd.read_csv('Nimda1.csv')
Slammer=pd.read_csv('Slammer1.csv')
Code_Red_I = pd.read_csv('Code_Red_I1.csv')


Datasets = [WannaCrypt, Nimda,Slammer, Moscow_blackout,Code_Red_I ]
names = ['WannaCrypt', 'Nimda','Slammer', 'Moscow_blackout','Code_Red_I' ]
start = ['2017-05-10', '2001-09-16', '2003-01-23', '2005-05-23', '2001-07-13']
for i in range (len(Datasets)): 
    
    dataset=Datasets[i][["NOA", "NOW", "NOANP", "NOWNP", "AAPL", "MAPL", "AUAPL", "NODA", "NODW", 
                         "NOIW", "AED", "MED", "IAT", "IGP packets", "EGP packets", "Incomplete packets", 
                         "packet size"]]
    index=pd.date_range(start=start[i], periods=dataset.NOA.count(), freq='min')
    dataset.insert(0, 'TimeStamp', index)
    dataset.tail()
    dataset.set_index('TimeStamp')
    month_average=dataset.resample('60min',on='TimeStamp').mean()
    month_average.head()

    plt.figure()
    month_average.plot(kind="line", figsize=(18, 9))
    plt.xlabel(names[i] + 'Dataset Collection Period', fontsize=12)
    plt.title('Trends of features', fontsize=16)
    #plt.ylabel('Unemployment Rate', fontsize=14)
    plt.grid(True)
    plt.legend(title="features",loc='upper right', fontsize=8, fancybox=True)

    plt.show()

```
